```python
#Sorting by last name (Sorting concept)

def display(student_records):
    print("\n******************************************")
    print("Welcome to your School Information Records!")
    print("*********************************************")
    
    print("Last Name, First Name Initial")
    print("-------------------------------------")
    
    for student in student_records:
        print(f"{student['last_name']}, {student['first_name']} {student['initial']}")
    
    print("\n******************************************")
    print("------------End of School Records------------")
    print("*********************************************")
    
def bubble_sort_by_last_name(student_records):
    n = len(student_records)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if student_records[j]['last_name'] > student_records[j + 1]['last_name']:
                student_records[j], student_records[j + 1] = student_records[j + 1], student_records[j]

def selection_sort_by_last_name(student_records):
    n = len(student_records)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if student_records[j]['last_name'] < student_records[min_index]['last_name']:
                min_index = j
                student_records[i], student_records[min_index] = student_records[min_index], student_records[i]

def insertion_sort_by_last_name(student_records):
    for i in range(1, len(student_records)):
        key = student_records[i]
        j = i - 1
        while j >= 0 and key['last_name'] < student_records[j]['last_name']:
            student_records[j + 1] = student_records[j]
            j -= 1
        student_records[j + 1] = key

def sort_by_last_name(student_records, sorting_choice):
    lower_case_algorithm = sorting_choice.lower()
    if lower_case_algorithm == 'bubble':
        bubble_sort_by_last_name(student_records)
    elif lower_case_algorithm == 'selection':
        selection_sort_by_last_name(student_records)
    elif lower_case_algorithm == 'insertion':
        insertion_sort_by_last_name(student_records)

student_records = []
num = int(input("How many students would you like to record?"))

for k in range(0, num):
    print("Enter the name of the student ")
    first_name = input("Enter first name: ")
    last_name = input("Enter last name: ")
    initial = input("Enter Middle Initial: ")
    
    student  = {
        'first_name': first_name,
        'last_name': last_name,
        'initial': initial,
    }
    student_records.append(student)
    
sorting_choice = input("Enter sorting of your choice (bubble/selection/insertion): ")

sort_by_last_name(student_records, sorting_choice)

display(student_records)
```

```python
#Grade Tracker: Matrix Concept

import numpy as np

def getscore():
    print("Enter Student Name: ")

    total_formative = np.zeros(7, dtype=int)
    total_summative = np.zeros(2, dtype=int)

    formative = np.zeros(7, dtype=int)
    summative = np.zeros(2, dtype=int)

    print("Formative Assessments: ")
    for x in range(7):
        formative[x] = int(input(f"Formative Assessment {x + 1}: "))
        total_formative[x] = int(input(f"Enter total number of points for Formative Assessment {x + 1}: "))

    print("\nSummative Assessments: ")
    for x in range(2):
        summative[x] = int(input(f"Summative Assessment {x + 1}: "))
        total_summative[x] = int(input(f"Enter total number of points for Summative Assessment {x + 1}: "))

    fa_percent = float(input("Enter the percentage for the formative assessments (e.g., 60 for 60%): ")) / 100
    sa_percent = float(input("Enter the percentage for the summative assessments (e.g., 40 for 40%): ")) / 100

    return fa_percent, sa_percent, formative, summative, total_formative, total_summative

def rate(ave):
    if ave >= 0 and ave < 50:
        return 'F'
    elif ave >= 50 and ave < 57:
        return 'D'
    elif ave >= 57 and ave < 64:
        return 'D+'
    elif ave >= 64 and ave < 71:
        return 'C'
    elif ave >= 71 and ave < 78:
        return 'C+'
    elif ave >= 78 and ave < 85:
        return 'B'
    elif ave >= 85 and ave < 92:
        return 'B+'
    elif ave >= 92 and ave <= 100:  
        return 'A'
    else:
        return 'Invalid'

def average(formative, summative, fa_percent, sa_percent, total_formative, total_summative):
    fa_ave = (np.sum(formative / total_formative) / len(formative)) * fa_percent * 100 
    sa_ave = (np.sum(summative / total_summative) / len(summative)) * sa_percent * 100 
    total_grade = fa_ave + sa_ave
    rounded_ave = round(total_grade, 2)
    return rounded_ave


def display(record):
    print("\n*********************************************************************************************************************")
    print("---------------------------------------------------Student Grade Tracker----------------------------------------------")
    print("-------------------------------------------------------S.Y 2023-2024-----------------------------------------------")
    print("-------------------------------------------------------1st Semester-----------------------------------------")
    print("**********************************************************************************************************************")
    print("\n")
    print("No.\tName\tQ1\tQ2\tQ3\tQ4\tQ5\tQ6\tQ7\tFinal1\tFinal2\tAverage\tRating")
    for i, row in enumerate(record):
        print(f"{i + 1}.\t{row[0]}\t{row[1]}\t{row[2]}\t{row[3]}\t{row[4]}\t{row[5]}\t{row[6]}\t{row[7]}\t{row[8]}\t{row[9]}\t{row[10]}\t{row[11]}")

max_students = int(input("Enter the maximum number of students: "))
student_data = []

for x in range(max_students):
    name = input("Enter Name: ")
    fa_percent, sa_percent, formative, summative, total_formative, total_summative = getscore()
    average_score = average(formative, summative, fa_percent, sa_percent, total_formative, total_summative)
    rating = rate(average_score)
    student_data.append([name] + formative.tolist() + summative.tolist() + [average_score, rating])

display(np.array(student_data))
```

```python
from collections import defaultdict, deque

def sort_students(students, struct, grade_ranges):
    sections = defaultdict(struct)
    
    for student in students:
        name, grade = student
        
        for sect, (lower, upper) in grade_ranges.items():
            if lower <= int(grade) <= upper:
                if struct is set:
                    sections[sect].add(name)
                else:
                    sections[sect].append(name)
                break
    return sections

def student_input():
    students = []
    
    while True:
        name = input("Enter student name, enter 'done' to finish input: ")
        if name.lower() == 'done':
            break
        grade = input("Enter grade of the student: ")
        students.append([name, grade])
    return students

print("1. sets")
print("2. stacks")
print("3. queues")

struct_pick = int(input("Choose a data structure to group your students: "))

if struct_pick == 1:
    struct = list
elif struct_pick == 2:
    struct = deque
elif struct_pick == 3:
    struct = set
else:
    print("Invalid input!")
    exit()

section_grade_ranges = {
    "A": (90, float('inf')),
    "B": (80, 89), 
    "C": (70, 79),
    "D": (-float('inf'), 70),
}

students = student_input()

sections = sort_students(students, struct, section_grade_ranges)

print("\n****Student Sections for S.Y 2023-2024****************\n")

for grade, students_in_section in sections.items():
    print(f"Section {grade}: {', '.join(students_in_section)}")
    

print("\n********End of Student Sections*************************\n")
```

```python
class TreeNode:
    def __init__(self, student_name):
        self.student_name = student_name
        self.left = None
        self.right = None

def insert_student(root, student_name):
    if root is None:
        return TreeNode(student_name)
    else:
        if student_name < root.student_name:
            root.left = insert_student(root.left, student_name)
        else:
            root.right = insert_student(root.right, student_name)
    return root

def inorder_traversal(root):
    if root:
        inorder_traversal(root.left)
        print(root.student_name)
        inorder_traversal(root.right)

def build_seat_arrangement(students):
    root = None
    for student in students:
        root = insert_student(root, student)
    return root

students = {"Paolo", "Naomi", "Kaye", "Marco", "Sam"}

seat_arrangement_root = build_seat_arrangement(students)
print("Seating Arrangement in Inorder Traversal:")
inorder_traversal(seat_arrangement_root)
```
